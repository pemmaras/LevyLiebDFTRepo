```python
import numpy as np
import pickle
import matplotlib
import matplotlib.pyplot as plt
```

```python
font = {'family' : 'Serif',
        'weight' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)

fontb = {'family' : 'Serif',
         'color'  : 'black',
        'weight' : 'normal',
        'size'   : 16}
font3 = {'family' : 'Serif',
        'weight' : 'bold',
        'size'   : 12}
font4 = {'family' : 'Serif',
        'weight' : 'normal',
        'size'   : 20}

colors = ['black', 'red', 'green', 'blue', 'purple']
```

```python
fname = 'dvs_U5.0_v1.0.pic'

with open(fname, 'rb') as fp:
    all_data = pickle.load(fp)
fp.close() 
fci = all_data['fci']
dvs = all_data['dvs']
opt_log = dvs['opt_data']

n_log = opt_log[:,0]
e_log = opt_log[:,1]
iter = np.arange(len(n_log))
n_fci = np.ones(len(n_log)) * fci['n_fci'][0]
plt.plot(iter, n_log, label='DVS')
plt.plot(iter, n_fci, '.', label='Exact')
plt.ylabel('$n_1$', fontdict=fontb)
plt.xlabel('Number of $F_{LL}[n_1]$ evaluations', fontdict=fontb)
plt.xlim([0,34])
# plt.ylim([-20,0])

plt.legend(prop=font)
```

```python
fname = 'dvs_U5.0_v1.0.pic'

with open(fname, 'rb') as fp:
    all_data = pickle.load(fp)
fp.close() 
fci = all_data['fci']
dvs = all_data['dvs']
opt_log = dvs['opt_data']

n_log = opt_log[:,0]
e_log = opt_log[:,1]
iter = np.arange(len(n_log))
n_fci = np.ones(len(n_log)) * fci['n_fci'][0]
n_err = np.abs(n_fci - n_log)
plt.plot(iter, n_err)
# plt.plot(iter, n_fci, '.', label='Exact')
plt.ylabel('$n_1~\mathrm{abs. error}$', fontdict=font4)
plt.xlabel('Number of $F_{LL}[n_1]$ evaluations', fontdict=font4)
plt.xticks(np.arange(0,35,5),fontsize=20)
plt.yticks([1e-4, 1e-3, 1e-2, 1e-1], fontsize=20)
plt.xlim([0,34])
plt.yscale('log')
# plt.xscale('')
# plt.ylim([-20,0])

# plt.legend(prop=font)
```
